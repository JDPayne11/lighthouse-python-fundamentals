#### 21. Consider a generator function that generates 10 integers and use it to build an array (★☆☆)

````python
def generator(integers):
    array = np.array([])
    integer = 0
    while integer < integers:
        number = np.random.randint(1,10)
        array = np.append(array,number)
        integer += 1
    return array

print(generator(10))
````

#### 30. Consider the vector \[1, 2, 3, 4, 5\], how to build a new vector with 3 consecutive zeros interleaved between each value? (★★★)

````python
vector = np.arange(1,6)
new_vector = np.zeros(3, dtype = 'int')
final = np.array([])
for i in range(5):
    final = np.append(final, vector[i])
    if i < 4:
        final = np.append(final, new_vector)

final = final.astype(np.int32)
print(final)

    
````

#### 34. Get the n largest values of an array (★★★)

````python
def largest_values(n):
    array = np.random.randint(0,10,20)
    sorted_array = np.sort(array)
    max_values = sorted_array[-n:]
    return max_values

print(largest_values(8))
````

#### 35. Extract unique rows from a two dimensional array? (★★★)

````python
def unique_row(n):
    array = np.random.uniform(0,8,size= (5,2))
    return array[[n]]

print(unique_row(3))
````
